```python
#since the salary csv only has abbreviation of team name but the supoerbowl csv has full name
#we have to create a dictionary to include the full team name for salary csv 
team_name = {'ARI': 'Arizona Cardinals',
'ATL': 'Atlanta Falcons',
'BAL': 'Baltimore Ravens',
'BUF': 'Buffalo Bills',
'CAR': 'Carolina Panthers',
'CHI': 'Chicago Bears',
'CIN': 'Cincinnati Bengals',
'CLE': 'Cleveland Browns',
'DAL': 'Dallas Cowboys',
'DEN': 'Denver Broncos',
'DET': 'Detroit Lions',
'GB': 'Green Bay Packers',
'HOU': 'Houston Texans',
'IND': 'Indianapolis Colts',
'JAX': 'Jacksonville Jaguars',
'KC': 'Kansas City Chiefs',
'LAC': 'Los Angeles Chargers',
'LAR': 'Los Angeles Rams',
'MIA': 'Miami Dolphins',
'MIN': 'Minnesota Vikings',
'NE': 'New England Patriots',
'NO': 'New Orleans Saints',
'NYG': 'New York Giants',
'NYJ': 'New York Jets',
'OAK': 'Oakland Raiders',
'PHI': 'Philadelphia Eagles',
'PIT': 'Pittsburgh Steelers',
'SEA': 'Seattle Seahawks',
'SF': 'San Francisco 49ers',
'TB': 'Tampa Bay Buccaneers',
'TEN': 'Tennessee Titans',
'WAS': 'Washington Redskins'}
```

```python
#here we are mapping the full name into the qb salary csv
qb_salary["Team Name"] = qb_salary['Team'].map(team_name)
```

```python
qb_salary.head()
```

```python
qb_salary.dtypes
```

```python
qb_salary['Signing.Bonus'].dtype
```

```python
#we want to turn the signing bonus into an interger but it has a dash that needs to be
#turned into a zero first
qb_salary['signing bonus'] = qb_salary['Signing.Bonus'].replace(to_replace = "-", value = 0)
```

```python
#check if the code works
qb_salary[qb_salary['signing bonus']==0].head()
```

```python
#now convert the column into an interger
qb_salary['signing bonus'] = qb_salary['signing bonus'].astype(int)
```

```python
qb_salary['signing bonus'].head()
```

```python
#we want to see total payment for the player that year so add salary + bonus
qb_salary['total cash'] = qb_salary['Base.Salary'] + qb_salary['signing bonus']
```

```python
qb_salary.head()
```

```python
superbowl.dtypes
```

```python
#conver year into proper format
qb_salary['Year'] = pd.to_datetime(qb_salary['Year'].astype(str), format='%Y').dt.year
qb_salary['Year'].head()
```

```python
#turn date into datetime
superbowl['Date'] = pd.to_datetime(superbowl['Date'])
superbowl['Date'].head()
```

```python
superbowl['Date'].dt
```

```python
#only show the year
superbowl['Date'] = superbowl['Date'].dt.year
superbowl.head()
```

```python
#taking the highest paid player per year
highest_paid_qb = qb_salary.groupby(['Year'])['total cash'].transform(max) == qb_salary['total cash']
qb_salary[highest_paid_qb]
```

```python
#formating the top paid qb by year
top_paid_qb = qb_salary[highest_paid_qb].sort_values('Year', ascending=True)
top_paid_qb
```

```python
results2 = pd.merge(top_paid_qb, superbowl, how='left',
        left_on='Year', right_on='Date')
```

```python
x = range(5)
# this is equal to x = [0, 1, 2, 3, ...9]
plt.bar(x, top_paid_qb['total cash']/ 1000)
plt.xticks(x, top_paid_qb['Player'], rotation='vertical') #show it vertical vs horizontally
plt.title('Top Paid QB in the NFL per Year (2012 to 2016)')
plt.ylabel('Total Cash')
plt.show()

print('2012 Superbowl Winner: ' + results2['QB  Winner'][results2.Year == 2012])
print('2013 Superbowl Winner: ' + results2['QB  Winner'][results2.Year == 2013])
print('2014 Superbowl Winner: ' + results2['QB  Winner'][results2.Year == 2014])
print('2015 Superbowl Winner: ' + results2['QB  Winner'][results2.Year == 2015])
print('2016 Superbowl Winner: ' + results2['QB  Winner'][results2.Year == 2016])
```

```python
def return_winner(year):
    print( str(year) + ' Superbowl winner: ' + results2['QB  Winner'][results2.Year == year])
    print( str(year+1) + ' Superbowl winner: ' + results2['QB  Winner'][results2.Year == year+1])
    print( str(year+2) + ' Superbowl winner: ' + results2['QB  Winner'][results2.Year == year+2])
    print( str(year+3) + ' Superbowl winner: ' + results2['QB  Winner'][results2.Year == year+3])
    print( str(year+4) + ' Superbowl winner: ' + results2['QB  Winner'][results2.Year == year+4])
```

```python
#creating a unique list of QBs that have won a superbowl and entering
#in the amount of superbowls that they won
def unique(list1): 
  
    # intilize a null list 
    unique_list = [] 
      
    # traverse for all elements 
    for x in list1: 
        # check if exists in unique_list or not 
        if x not in unique_list: 
            unique_list.append(x) 
    # print list 
    for x in unique_list: 
        print(x)
```

```python
#applying the function
superbowl_count = (superbowl['QB  Winner'])
#superbowl_count = unique(list1)
#superbowl_count
superbowl_count
```

```python
#putting the above into a table and formating the columns (sort by names) 
superbowl_count = superbowl.groupby(by='QB  Winner', as_index=False).agg({'Date': pd.Series.nunique})
superbowl_count
```

```python
#more formatting..changing Date to number of superbowls
superbowl_count.rename(columns={'Date':'Number.of.superbowls'}, inplace=True)
superbowl_count
```

```python
qb_salary.head()
```

```python
qb_salary_2016 = qb_salary[qb_salary.Year == 2016]
qb_salary_2016.head()
```

```python
salary_superbowl = pd.merge(qb_salary_2016, superbowl_count, how='left',
        left_on='Player', right_on='QB  Winner')
salary_superbowl.head()
```

```python
#chaning the Nan to 0
salary_superbowl['Number.of.superbowls'].fillna(0, inplace=True)
salary_superbowl.head()
```

# Superbowl Rings vs QB Total Payment

```python
data_25 = salary_superbowl.groupby('Number.of.superbowls').quantile(0.25)
data_75 = salary_superbowl.groupby('Number.of.superbowls').quantile(0.75)
data_median = salary_superbowl.groupby('Number.of.superbowls').median()

#top_paid_2016 = salary_superbowl.sort_values('total cash', ascending = False).head(10)

#data_25 = salary_superbowl.sort_values('total cash', ascending = False).quantile(0.25)
#data_75 = salary_superbowl.sort_values('total cash', ascending = False).quantile(0.75)
#data_median = salary_superbowl.sort_values('total cash', ascending = False).median()

plt.plot(data_75.index, data_75['total cash']/ 1000, 'green')
plt.plot(data_median.index,  data_median['total cash']/ 1000, 'red')
plt.plot(data_25.index, data_25['total cash']/ 1000, 'blue')
plt.plot(salary_superbowl['Number.of.superbowls'],  salary_superbowl['total cash']/ 1000, 'o', markersize = 2, alpha = 0.9)
plt.title('Salary vs Number of Rings')
plt.ylabel('Total Cash')
plt.xlabel('Superbowl Rings')

plt.legend(['75th Percentile', 'Median', '25th Percentile'])
plt.show()
```
